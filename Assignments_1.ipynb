```python
soup = BeautifulSoup(page.content,"html.parser")
print(soup.prettify())
```

```python
#scrap president names
president_names = soup.find_all('div',class_="presidentListing")
president_names
```

```python
# parse president names
presidents = []

for president in president_names:
    president = president.text.split('\n')[1]
    president = president.strip("")
    presidents.append(president)
    
presidents
```

```python
# parse president names
Term_office=[]

for Term in president_names:
    Term = Term.text.split('\n')[2]
    Term = Term.strip("")
    Term_office.append(Term)
    
Term_office
```

```python
import pandas as pd

import warnings
warnings.filterwarnings('ignore')
df=pd.DataFrame()
df['President Names'] = presidents
df['Term of office'] = Term_office
df.head()
```
